```python
%matplotlib notebook
import matplotlib as mpl
import matplotlib.pyplot as plt
from matplotlib import dates
import seaborn as sn
import os
import re
import nibabel as nib
import numpy as np
import pandas as pd
from shutil import copyfile
from datetime import datetime
from copy import deepcopy
import time
import ssl
import requests
from bs4 import BeautifulSoup
plt.rcParams.update({'font.size': 16})
```

```python
website = 'https://www.mfiles.co.uk/'
url = os.path.join(website, 'classical-midi.htm')
page = requests.get(url, verify=False)
soup = BeautifulSoup(page.text, 'html.parser')
import ssl
ssl._create_default_https_context = ssl._create_unverified_context
```

```python
main_path = 'mfiles'
for child in soup.find_all('li'):
    link = child.find('a').get('href')
    if link.endswith('.mid'):
        mid_url = os.path.join(website, link)
        print(mid_url)
        response = requests.get(mid_url)
        path = os.path.join(main_path, link.split('/')[-1])
        open(path, 'wb').write(response.content)
```

```python
website = 'https://www.midiworld.com/'
url = os.path.join(website, 'classic.htm')
page = requests.get(url, verify=False)
soup = BeautifulSoup(page.text, 'html.parser')
import ssl
ssl._create_default_https_context = ssl._create_unverified_context
```

```python
main_path = 'midiworld'
for child in soup.find_all('a'):
    link = child.get('href')
    if link is not None and link.endswith('.htm'):
        mid_page = requests.get(os.path.join(website, link), verify=False)
        mid_soup = BeautifulSoup(mid_page.text, 'html.parser')
        for mid_child in mid_soup.find_all('a'):
            mid_link = mid_child.get('href')
            if link is not None and mid_link.endswith('.mid'):
                print(mid_link)
                response = requests.get(mid_link)
                path = os.path.join(main_path, mid_link.split('/')[-1])
                open(path, 'wb').write(response.content)
    elif link is not None and link.endswith('.mid'):
        print(link)
        response = requests.get(link)
        path = os.path.join(main_path, link.split('/')[-1])
        open(path, 'wb').write(response.content)
```
